```python
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import pandas as pd # panda first

#https://github.com/twopirllc/pandas-ta#features
# indicators library for technical anlysis
import pandas_ta as ta # import 2nd


df = pdr.get_data_yahoo('XOM','2020-01-01','2020-12-11')

df_sma = ta.sma(df['Close'],length=14 )
df_rsi = ta.rsi(df['Close'],length=25 )

plt.subplot(2,1,1)
plt.grid()
plt.title('Price at Close')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.plot(df['Close'],label='XOM')
plt.plot(df_sma,label='SMA')
plt.legend()

plt.subplot(2,1,2)
plt.grid()
plt.title('RSI')
plt.plot(df_rsi,label='RSI')
plt.legend()

plt.show()
```
